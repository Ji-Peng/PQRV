```python
def expand_macro(macro_name, *args):
    if macro_name == "xoror":
        template = """or  {T}, {S01}, {S02}
xor {out}, {S00}, {T}"""
    elif macro_name == "xornotor":
        template = """not {T}, {S01}
or  {T}, {T}, {S02}
xor {out}, {S00}, {T}"""
    elif macro_name == "xorand":
        template = """and {T}, {S01}, {S02}
xor {out}, {S00}, {T}"""
    elif macro_name == "xorornot":
        template = """not {T}, {S02}
or  {T}, {T}, {S01}
xor {out}, {S00}, {T}"""
    elif macro_name == "xornotand":
        template = """not {T}, {S01}
and {T}, {T}, {S02}
xor {out}, {S00}, {T}"""
    elif macro_name == "notxoror":
        template = """not {T0}, {S00}
or  {T}, {S01}, {S02}
xor {out}, {T0}, {T}"""
    elif macro_name == "notxorand":
        template = """not {T0}, {S00}
and {T}, {S01}, {S02}
xor {out}, {T0}, {T}"""
    else:
        return "Macro not recognized."

    # Unpack the arguments
    if macro_name in ["notxoror", "notxorand"]:
        out, S00, S01, S02, T, T0 = args
        expanded = template.format(out=out, S00=S00, S01=S01, S02=S02, T=T, T0=T0)
    else:
        out, S00, S01, S02, T = args
        expanded = template.format(out=out, S00=S00, S01=S01, S02=S02, T=T)
    
    return expanded

# List of macro calls to expand
macro_calls = [
    ("xoror", "\\S05", "\\S10", "\\S11", "\\S07", "\\T04"),
    ("xorand", "\\S06", "\\S11", "\\S07", "\\S08", "\\T03"),
    ("xorornot", "\\S07", "\\S07", "\\S08", "\\S09", "\\T02"),
    ("xoror", "\\S08", "\\S08", "\\S09", "\\S10", "\\T04"),
    ("xorand", "\\S09", "\\S09", "\\S10", "\\S11", "\\T03"),

    ("xoror", "\\S10", "\\S15", "\\S16", "\\S12", "\\T04"),
    ("xorand", "\\S11", "\\S16", "\\S12", "\\S13", "\\T03"),
    ("xornotand", "\\S12", "\\S12", "\\S13", "\\S14", "\\T02"),
    ("notxoror", "\\S13", "\\S13", "\\S14", "\\S15", "\\T04", "\\T03"),
    ("xorand", "\\S14", "\\S14", "\\S15", "\\S16", "\\T02"),

    ("xorand", "\\S15", "\\S20", "\\S21", "\\S17", "\\T04"),
    ("xoror", "\\S16", "\\S21", "\\S17", "\\S18", "\\T03"),
    ("xornotor", "\\S17", "\\S17", "\\S18", "\\S19", "\\T02"),
    ("notxorand", "\\S18", "\\S18", "\\S19", "\\S20", "\\T04", "\\T03"),
    ("xoror", "\\S19", "\\S19", "\\S20", "\\S21", "\\T02"),

    ("xornotand", "\\S20", "\\S00", "\\S01", "\\S22", "\\T04"),
    ("notxoror", "\\S21", "\\S01", "\\S22", "\\S23", "\\T03", "\\T02"),
    ("xorand", "\\S22", "\\S22", "\\S23", "\\S24", "\\T04"),
    ("xoror", "\\S23", "\\S23", "\\S24", "\\S00", "\\T03"),
    ("xorand", "\\S24", "\\S24", "\\S00", "\\S01", "\\T02"),

    ("xoror", "\\S00", "\\T00", "\\T01", "\\S02", "\\T04"),
    ("xornotor", "\\S01", "\\T01", "\\S02", "\\S03", "\\T03"),
    ("xorand", "\\S02", "\\S02", "\\S03", "\\S04", "\\T02"),
    ("xoror", "\\S03", "\\S03", "\\S04", "\\T00", "\\T04"),
    ("xorand", "\\S04", "\\S04", "\\T00", "\\T01", "\\T03")
]

# Expand each macro call and print the result
for macro_call in macro_calls:
    macro_name, *arguments = macro_call
    result = expand_macro(macro_name, *arguments)
    print(result)
    print()  # Add a blank line for readability
```

```python
def expand_XOR5(out, S00, S01, S02, S03, S04):
    instructions = [
        f"xor {out}, {S00}, {S01}",
        f"xor {out}, {out}, {S02}",
        f"xor {out}, {out}, {S03}",
        f"xor {out}, {out}, {S04}"
    ]
    return instructions

# 示例用法
macro_calls = [
    ("\\T00", "\\S00", "\\S05", "\\S10", "\\S15", "\\S20"),
    ("\\T01", "\\S02", "\\S07", "\\S12", "\\S17", "\\S22"),
    ("\\T03", "\\S01", "\\S06", "\\S11", "\\S16", "\\S21"),
    ("\\T02", "\\S04", "\\S09", "\\S14", "\\S19", "\\S24"),
    ("\\T04", "\\S03", "\\S08", "\\S13", "\\S18", "\\S23")
]

expanded_instructions = []

for call in macro_calls:
    out, S00, S01, S02, S03, S04 = call
    expanded = expand_XOR5(out, S00, S01, S02, S03, S04)
    expanded_instructions.extend(expanded)

# 输出展开后的指令
for instr in expanded_instructions:
    print(instr)
```

```python
def expand_ROLn(out, inp, tmp, n):
    return [f"rori {out}, {inp}, {64-n}"]

def expand_ROL1(out, inp, tmp):
    return expand_ROLn(out, inp, tmp, 1)

# 示例用法
macro_calls = [
    ("\\T02", "\\T01", "\\T03"),
    ("\\T00", "\\T00", "\\T01"),
    ("\\T04", "\\T04", "\\T01"),
    ("\\T03", "\\T03", "\\T01"),
    ("\\T01", "\\S06", "\\T02", "44"),
    ("\\S00", "\\S02", "\\T03", "62"),
    ("\\S02", "\\S12", "\\T02", "43"),
    ("\\S12", "\\S13", "\\T03", "25"),
    ("\\S13", "\\S19", "\\T02", "8"),
    ("\\S19", "\\S23", "\\T03", "56"),
    ("\\S23", "\\S15", "\\T02", "41"),
    ("\\S15", "\\S01", "\\T03", "1"),
    ("\\S01", "\\S08", "\\T02", "55"),
    ("\\S08", "\\S16", "\\T03", "45"),
    ("\\S16", "\\S07", "\\T02", "6"),
    ("\\S07", "\\S10", "\\T03", "3"),
    ("\\S10", "\\S03", "\\T02", "28"),
    ("\\S03", "\\S18", "\\T03", "21"),
    ("\\S18", "\\S17", "\\T02", "15"),
    ("\\S17", "\\S11", "\\T03", "10"),
    ("\\S11", "\\S09", "\\T02", "20"),
    ("\\S09", "\\S22", "\\T03", "61"),
    ("\\S22", "\\S14", "\\T02", "39"),
    ("\\S14", "\\S20", "\\T03", "18"),
    ("\\S20", "\\S04", "\\T02", "27"),
    ("\\S04", "\\S24", "\\T03", "14"),
    ("\\S24", "\\S21", "\\T02", "2"),
    ("\\S21", "\\S05", "\\T03", "36")
]

expanded_instructions = []

for call in macro_calls:
    if len(call) == 3:
        out, inp, tmp = call
        expanded = expand_ROL1(out, inp, tmp)
    elif len(call) == 4:
        out, inp, tmp, n = call
        expanded = expand_ROLn(out, inp, tmp, int(n))
    else:
        continue
    expanded_instructions.extend(expanded)

# 输出展开后的指令
for instr in expanded_instructions:
    print(instr)
```
