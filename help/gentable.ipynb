```python
# for 7-layer NTT
def bit_reverse(input,bit_number):
    output=0
    for i in range(bit_number):
        # get i-th bit
        bit=(input>>i)&1
        output+=bit<<(bit_number-1-i)
    return output

def gen_order_ntt(in_list, layers):
    out_list=[]
    for i in in_list:
        for j in range(0,layers):
            # constant numbers of each layer
            num=1<<j
            for k in range(0,num):
                out_list.append(i*num+k)
    return out_list

def gen_order_intt(in_list, layers):
    out_list=[]
    for i in in_list:
        for j in range(layers-1,-1,-1):
            # print(j)
            # constant numbers of each layer
            num=1<<j
            for k in range(0,num):
                out_list.append(i*num+k)
    return out_list

tree_128=[]
for i in range(1,128):
    tree_128.append(bit_reverse(i,7))
print("power of zeta & NTT")
print(tree_128)

tree_128_inv=[]
for i in range(6,-1,-1):
    const_number=1<<i
    for j in range(const_number,const_number*2):
        tree_128_inv.append(bit_reverse(j,7))
print("power of zeta & INTT")
print(tree_128_inv,"len =",len(tree_128_inv))

# 1-layer merging
in_list1=[1]
out_list1=in_list1
# 6-layer merging
in_list2=[2,3]
out_list2=gen_order_ntt(in_list2,6)
out_list=out_list1+out_list2
print("out_list: ",out_list, "len=",len(out_list))
br_out_list=[]
for i in out_list:
    br_out_list.append(bit_reverse(i,7))
print("power of zeta & NTT & 1+6 layers merging")
print(br_out_list)

in_list_intt1=[2,3]
in_list_intt2=[1]
out_list_intt1=gen_order_ntt(in_list_intt1,6)
out_list_intt2=in_list_intt2
out_list_intt=out_list_intt1+out_list_intt2
br_out_list_intt=[]
for i in out_list_intt:
    br_out_list_intt.append(bit_reverse(i,7))
print("Structural order & INTT & layers merging")
print(br_out_list_intt)

out_list_intt1=gen_order_intt(in_list_intt1,6)
out_list_intt2=in_list_intt2
out_list_intt=out_list_intt1+out_list_intt2
print("out_list_intt", out_list_intt, "len=", len(out_list_intt))
br_out_list_intt=[]
for i in out_list_intt:
    br_out_list_intt.append(bit_reverse(i,7))
print("Usage order & INTT & 1+6 layers merging")
print(br_out_list_intt)

# point-wise multiplication: mod (X^2-zeta^{2br(i)+1})
mul_table_tree=[]
for i in range(128):
    mul_table_tree.append(2*bit_reverse(i,7)+1)
print("(X^2-zeta^{2br(i)+1})")
print(mul_table_tree)

# def gen_order_pointwise(inlist):
#     out_list = []
#     for i in inlist:
#         for j in range(0, 8):
#             out_list.append(i+j*2)
#     return out_list
# # point-wise multiplication for special ntt order
# in_list_pointwise = [0, 1, 16, 17, 32, 33, 48, 49, 64, 65, 80, 81, 96, 97, 112, 113]
# out_list_pointwise = gen_order_pointwise(in_list_pointwise)
# print("out list for point-wise multiplication")
# print(out_list_pointwise)
# mul_table_tree = []
# for i in out_list_pointwise:
#     mul_table_tree.append(2*bit_reverse(i,7)+1)
# print("(X^2-zeta^{2br(i)+1})")
# print(mul_table_tree)
```
