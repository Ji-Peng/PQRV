```python
items_mont_input=[]
intt_range=[[],[],[],[],[],[],[],[],[],[]]
mont_output='(-Q,Q)'
N=256

layer=0
for i in range(0, N):
    intt_range[layer].append("a_%d" % (i))
print('layer 0: ', intt_range[layer])

for layer in range(1, 9):
    for i in range(0, N):
        intt_range[layer].append('')

# layer=1
# len=1
# for start in range(0, N, len):
#     for j in range(start, N, len):
#         print(j, j+len)
#         intt_range[layer][j]=intt_range[layer-1][j]+'+'+intt_range[layer-1][j+len]
#         intt_range[layer][j+len]=mont_output
#         items_mont_input.append(intt_range[layer-1][j]+'-'+intt_range[layer-1][j+len])
# print('layer 1: ', intt_range[layer])

layer=1
len=1
while len < N:
    start=0
    while start < N:
        j=start
        while j<start+len:
            # print(j, j+len)
            intt_range[layer][j]=intt_range[layer-1][j]+'+'+intt_range[layer-1][j+len]
            intt_range[layer][j+len]=mont_output
            items_mont_input.append(intt_range[layer-1][j]+'-'+intt_range[layer-1][j+len])
            j+=1
        # update start
        start=j+len
    # update len
    len=len<<1
    print('layer %d: ' % (layer), intt_range[layer])
    layer+=1

print('items_mont_input: ', items_mont_input)
```
